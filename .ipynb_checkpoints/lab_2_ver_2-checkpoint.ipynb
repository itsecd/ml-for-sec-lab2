```python
def noise(n=131072):# n - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —á–∏—Å–µ–ª, r - –ø—Ä–æ—Å—Ç–æ —á–∏—Å–ª–æ –¥–ª—è –ø—Å–µ–≤–¥–æ—Ä–∞–Ω–¥–æ–º–∞
    b=[random.randint(0,1) for i in range(0,n)]
    return b

def random_numbers(n):# n - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —á–∏—Å–µ–ª, r - –ø—Ä–æ—Å—Ç–æ —á–∏—Å–ª–æ –¥–ª—è –ø—Å–µ–≤–¥–æ—Ä–∞–Ω–¥–æ–º–∞
  
    b=[random.choice([-1,1]) for i in range(0,n)]
    return b

def random_loc(n):# n - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —á–∏—Å–µ–ª, r - –ø—Ä–æ—Å—Ç–æ —á–∏—Å–ª–æ –¥–ª—è –ø—Å–µ–≤–¥–æ—Ä–∞–Ω–¥–æ–º–∞
    b=[random.randint(0,262144-1) for i in range(0,n)]
    return b
```

```python
def int_in_name(c)->str:
    while(len(c)!=5):
        c='0'+c
    return 'Data_for_test/Image'+c+'.tif'
        


def add_noise(k,image,n,location,rand_n):
    '''r - —Ä–∞–Ω–¥–æ–º–Ω–æ –¥–ª—è –≤—Å–µ—Ö –¥–ª—è –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ 1 –∏ -1
    image - –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ
    k - –∫–ª—é—á
    n - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–∑–º–µ–Ω—è–µ–º—ã—Ö –ø–∏–∫—Å–µ–ª–µ–π'''
    
    #—Å–æ–∑–¥–∞–µ–º —à—É–º –¥–ª—è –¥–∞–Ω–Ω–æ–≥–æ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è
    b=noise(n)
    
    #–ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –≤ –æ–¥–Ω–æ–º–µ—Ä–Ω—ã–π –º–∞—Å—Å–∏–≤
    image_one = image.flatten()
    
    f={0:1,255:254}
    
    for l in range(len(location)):
        loc=location[l]
        img=image_one[loc]
    
        if(img==0 or img==255):
            img=f[img]
        elif(bin2(img)!=b[l]):
            img+=rand_n[l]
        image_one[loc]=img

    return np.reshape(image_one, imarray.shape) # –ò–ª–∏ –Ω–µ—Ç. –î–≤—É–º–µ—Ä–Ω–æ–µ
#–í–æ—Ç —ç—Ç–æ –¥–ª—è –ø–æ–¥—Å—Ç—á—ë—Ç–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –≤—Å–µ–≥–æ –∏ —Ç–∞–∫ –º–æ–∂–Ω–æ –∏–º–µ–Ω–∞ –≤—Å–µ—Ö —Ñ–∞–π–ª–æ–≤ –ø–æ–ª—É—á–∏—Ç—å pictures = os.listdir(dir_name)
```

```python
#–º–µ—Ç–æ–¥ —Ä–∞–∑–≤–µ—Ä—Ç–∫–∏ –¥–≤—É–º–µ—Ä–Ω–æ–π –ø–ª–æ—Å–∫–æ—Å—Ç–∏ (—Å–µ—Ä–ø–∞–Ω—Ç–∏–Ω–Ω–∞—è)
def scan(image):
    n1,n2=image.shape
    image_scan=np.zeros(n1*n2,dtype=int)
    
    for i in range(0,n1):
        first=i*n2
        second=(i+1)*n2
        if(i%2==0):
            image_scan[first:second]=image[i]
        else:
            image_scan[first:second]=image[i][::-1]
    return image_scan
```

```python
#—Ä–∞—Å—Å—á—ë—Ç –¥–ª–∏–Ω—ã —Å–µ—Ä–∏–π —Ä–∞–∑–ª–∏—á–Ω–æ–π –¥–ª–∏–Ω—ã (–Ω–æ–ª–µ–π)
# –Ω—É–∂–Ω–æ —Ä–µ—à–∏—Ç—å –∫–∞–∫–∞—è –º–∞–∫—Å–∏–º–∞–ª—å–Ω–∞ –¥–ª–∏–Ω–∞ —Ä–µ—Å–∏–∏ –±—É–¥–µ—Ç (–ø—É—Å—Ç—å 20)
def c_of_series(image_scan_b1,z):
    res = ([len(list(g)) for i,g in groupby(image_scan_b1) if i==0])
    dict_s={}
    for i in res:
        if(i<z+1):
            if i in dict_s:
                dict_s[i]+=1
            else:
                dict_s[i]=1
    ll=[]
    for i in range(1,z+1):
        if i in dict_s:
            ll.append(dict_s[i])
        else:
            ll.append(0)
    return ll
    
```

```python
#–ø–æ—á–µ–º—É-—Ç–æ –º–µ–¥–ª–µ–Ω–Ω–µ–µ —Ä–∞–±–æ—Ç–∞–µ—Ç –∏–ª–∏ –Ω–µ—Ç
#—Ä–∞—Å—Å—á—ë—Ç –¥–ª–∏–Ω—ã —Å–µ—Ä–∏–π —Ä–∞–∑–ª–∏—á–Ω–æ–π –¥–ª–∏–Ω—ã (–Ω–æ–ª–µ–π)
# –Ω—É–∂–Ω–æ —Ä–µ—à–∏—Ç—å –∫–∞–∫–∞—è –º–∞–∫—Å–∏–º–∞–ª—å–Ω–∞ –¥–ª–∏–Ω–∞ —Ä–µ—Å–∏–∏ –±—É–¥–µ—Ç (–ø—É—Å—Ç—å 20)
def c_of_series(image_scan_b1,z):
    res = ([len(list(g)) for i,g in groupby(image_scan_b1) if i==0])
    dict_s=np.zeros(z,dtype=int)
    #dict_s=dict.fromkeys(range(1,21),0)
    for i in res:
        if(i<z+1):
            dict_s[i-1]+=1
    return list(dict_s)# –º–æ–∂–Ω–æ –Ω–∞–≤–µ—Ä–Ω–æ–µ —É–±—Ä–∞—Ç—å –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ —Å —Å–ø–∏—Å–æ–∫
    
```

```python
# –º–µ—Ç–æ–¥ –¥–ª—è –ø–µ—Ä–≤—ã—Ö K/2 –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π, –¥–æ–±–∞–ª—è–µ–º —à—É—Ç –∏ –ø–æ–ª—É—á–∞–µ–º —Å–µ—Ä–∏–∏
def add_noise_to_half_of_images(name_dir,k,n,z):# n –º–æ–∂–Ω–æ –∑–∞–º–µ–Ω–∏—Ç—å –Ω–∞ –ø—Ä–æ—Ü–µ–Ω—Ç—ã, –∞ —Ç–æ —Ç–∞–∫ –ø–æ–ª—É—á–∞–µ—Ç—Å—è –º—ã –æ–±—è–∑–∞–Ω—ã –∑–Ω–∞—Ç—å —Å–∫–æ–ª—å–∫–æ –∏—Ö —Ç–∞–º
    files = os.listdir(path=name_dir)
    K2=int(len(files)/2)
    sp_s=np.zeros((K2,z),dtype=int)
    
    #–≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –ø—Å–µ–≤–¥–æ—Å–ª—É—á–∞—Ñ–Ω—ã—Ö —á–∏—Å–µ–ª
    random.seed(k)
    rand_n=random_numbers(n)
    
    #–≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º –º–µ—Å—Ç–æ–ø–æ–ª–æ–∂–µ–Ω–∏–µ –ø–∏–∫—Å–µ–ª–µ–π
    location=random_loc(n)
    
    for i in range(K2):
        #—Å—á–∏—Ç—ã–≤–∞–µ–º –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ
        img = imread(name_dir+files[i])
        
        #–¥–æ–±–∞–≤–ª—è–µ–º —à—É–º
        noise_img=add_noise(k,img,n,location,rand_n)
        
        #—Å–µ—Ä–ø–∞–Ω—Ç–∏–Ω–Ω–∞—è —Ä–∞–∑–≤–µ—Ä—Ç–∫–∞
        image_scan=scan(noise_img)
        
        #–ø–æ–ª—É—á–µ–Ω–∏–µ 1 –ø–ª–æ—Å–∫–æ—Å—Ç–∏
        image_scan_b1=list(map(bin2,image_scan))
        
        #—Ä–∞—Å—Å—á—ë—Ç —Å–µ—Ä–∏–π
        sp_s[i][:]=c_of_series(image_scan_b1,z)
        print(i)
        
        
    return sp_s
    
```

```python
def test_and_train(K,df,y):
    n=int(K/2*0.5)
    a=df.head(n)
    b=df.tail(n)
    train_X=pd.concat([a,b],ignore_index=True)
    
    test_X=df.iloc[:-n].iloc[n:]
    test_X.reset_index(drop= True , inplace= True )
    
    train_y=y[:n]+y[-n:]
    
    test_y=y[n:-n]
    return train_X,test_X,train_y,test_y
```
