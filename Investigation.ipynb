```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
```

```python
x = np.random.randint(0, 2, size=99600, dtype='int32')
```

```python
im = cv2.imread('data/dog.jpg')[..., 0]
```

```python
def get_1_binary_plane(im):
    im_flat = im.flatten()
    plane = []
    for _x in im_flat:
        plane.append(np.bitwise_and(_x, 1))
    return np.array(plane, dtype='int32').reshape(im.shape)
```

```python
x2 = get_1_binary_plane(im)
```

```python
fft = np.fft.fft2(im[64:96, 64:96])
fft2 = np.fft.fft2(x.reshape(im.shape)[:32, :32])
```

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.heatmap(np.log(np.abs(fft)))
plt.subplot(1, 2, 2)
sns.heatmap(np.log(np.abs(fft2)))
```

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(np.log(np.abs(fft)).mean(axis=1))
plt.subplot(1, 2, 2)
plt.plot(np.log(np.abs(fft2)).mean(axis=1))
```

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(np.log(np.abs(fft)).mean(axis=0))
plt.subplot(1, 2, 2)
plt.plot(np.log(np.abs(fft2)).mean(axis=0))
```

```python
diff_fft = np.abs(np.log(np.abs(fft)) - np.log(np.abs(fft2)))
```
